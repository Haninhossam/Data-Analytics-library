```python
a = np.array([1,2,3,4])
b= np.array([[10],[20],[30]])
c =a+b
print(c)
# shape will be (3,4) because:
# b (3,1) broadcasts across columns to match (3,4)
# a (4,) broadcasts across rows to match (3,4)
```

```python
data = np.array([150,220,90,300,180,45])
n = (data - data.min())/( data - data.max())
print(n)
```

```python
z_scores = (data - data.mean()) / data.std()
print(z_scores)
```

```python
mat = np.array([[4,8,1,12],
                [7,3,9,5],
                [11,2,6,10],
                [15,0,14,13]])
row_m = mat.mean(axis=1 , keepdims= True)
print(row_m)
```

```python
row_S = mat.sum(axis=0 , keepdims= True)
print(row_S)
```

```python
max = mat.max()
print(max)
```

```python
p = np.unravel_index(mat.argmax() , mat.shape)
print(p)
```
