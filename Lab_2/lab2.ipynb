Load the file students.csv and make sure Score Date is parsed as datetime. 

```python
import pandas as pd
df = pd.read_csv("students.csv", parse_dates =["Score Date"])
```

•Print the shape, columns list, and data types of the DataFrame. 

```python
print("Shape:", df.shape)
print(df.columns.to_list())
print(df.dtypes)
```

Show the first 7 rows and the last 3 rows.

```python
print(df.head(7))
print(df.tail(3))
```

•Show how many missing values in each column.

```python
print(df.isna().sum())
```

•What is the average, minimum, and maximum Grade? 

```python
print(df['Grade'].mean())
print(df['Grade'].min())
print(df['Grade'].max())
```

•How many unique cities are there? List them. 

```python
print(df['City'].nunique())
print(df['City'].unique())
```

•Use value_counts() to show how many students per Subject. 

```python
print(df['Subject'].value_counts())
```

•Select only the columns: Name, Grade, Subject, City. 

```python
print(df[["Name", "Grade", "Subject", "City"]])
```

•Show all students who Passed and have Grade ≥ 85. 

```python
print(df[(df['Passed']== True) & (df['Grade'] >= 85)])
```

•Show students whose Name contains the letter "a" or "A" (case insensitive). 

```python
print(df[df['Name'].str.contains("a" ,case= False , na= False)])
```

Show students in Cairo or Alexandria with Grade < 80.

```python
print(df[df['City'].isin(["Cairo", "Alexandria"]) & (df["Grade"] < 80)])
```

•Show students with Grade between 80 and 90 (inclusive). 

```python
print(df[df["Grade"].between(80,90,inclusive="both")])
```

•Using .loc, change Khaled's grade from 55 to 68

```python
df.loc[(df["Name"] == "Khaled") & (df["Grade"] == 55) ,"Grade"] =68
print(df[df["Name"] == "Khaled"])
```

•Add a new column Is_Excellent that is True if Grade ≥ 90, False otherwise (vectorized). 

```python
df["is_excellent"] = df["Grade"] >= 90
df.head
```

•Fill missing Grade values with the average Grade. 

```python
df["Grade"] = df["Grade"] .fillna(df["Grade"].mean())
```

Fill missing Score Date values with the most common date (mode)

```python
df["Score Date"] = df["Score Date"] .fillna(df["Score Date"].mode()[0])
```

•Add a new column Final_Grade = Grade + 5 (bonus points). 


```python
df["Grade"] = df["Grade"] +5
```

•Create a column Grade_Category using pd.cut with bins:
•0–69 → "Fail"
•70–79 → "Average"
•80–89 → "Good"
•90–100 → "Excellent"

```python
df["grad_category"] = pd.cut(
  df["Grade"],
  bins=[0,69, 79 ,89,100],
  labels=["Fail", "Average", "Good", "Excellent"],
  include_lowest= True
)
```

```python
df.head()
```

•Remove the column Student ID. 

```python
df = df.drop(columns=["Student ID"])
```

•Keep only students who Passed. 

```python
df_passed = df[df["Passed"] == True]
```

•Sort the DataFrame by Grade descending. 


```python
df_sorted = df_passed.sort_values(by="Grade", ascending= False)
print(df_sorted)
```

•Sort by City ascending, then inside each city sort by Grade descending. 

```python
df_sorted_city = df_passed.sort_values(by=["City","Grade"] , ascending= [True ,False])
print(df_sorted_city)
```

•Add a column Rank that ranks students by Grade (1 = highest, handle ties with 'min'). 

```python
df_passed["rank"] = df_sorted["Grade"].rank(method="min",ascending= False)
print(df_passed)
```

•Show the top 4 students with the highest grades (use nlargest). 

```python
top_4 =df_passed.nlargest(4,"Grade")
print(top_4)
```

Show the 3 youngest students (use nsmallest on Age). 

```python
young = df_passed.nsmallest(3,"Age")
print(young)
```

•Find the student(s) with the highest Grade and show their Name and Grade.

```python
highest_grad = df_passed.loc[
  df_passed["Grade"] == df_passed["Grade"].max(),
  ["Name" ,"Grade"]]

print(highest_grad)
```

•After filling missing grades, recalculate the average Grade per Subject. 

```python
avg_per_subject =df.groupby("Subject")["Grade"].mean()
print(avg_per_subject)
```

•Create a new column City_Avg that shows the average grade of each student's city (using transform). 

```python
df["city_avg"] =df.groupby("City")["Grade"].transform("mean")
print(df.head())
```

•Show the subject with the highest average grade. 

```python
highest_avg = avg_per_subject.idxmax()
print("Subject:", highest_avg)
```
